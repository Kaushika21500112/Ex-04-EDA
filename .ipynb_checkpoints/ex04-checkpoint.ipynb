```python
import pandas as pd
import numpy as np
import seaborn as sns
df=pd.read_csv("supermarket.csv")
df
```

```python
#cleqaning dataset
#checking for null values
df.isnull().sum()
```

```python
#data is cleaned, proceed to remove outliers
#checking for outliers
df.boxplot()
```

```python
#outliers only involves numerical data
#removing all non numerical and alphanumerical data
df1=df.copy()
df1.drop('Invoice ID',axis=1,inplace=True)
df1.drop('Branch',axis=1,inplace=True)
df1.drop('City',axis=1,inplace=True)
df1.drop('Customer type',axis=1,inplace=True)
df1.drop('Gender',axis=1,inplace=True)
df1.drop('Product line',axis=1,inplace=True)
df1.drop('Date',axis=1,inplace=True)
df1.drop('Time',axis=1,inplace=True)
df1.drop('Payment',axis=1,inplace=True)
```

```python
df1
```

```python
#checking for outliers
df1.boxplot()
```

```python
#removing outliers
df2=df1.copy()
q1=df2.quantile(0.25)
q3=df2.quantile(0.75)
IQR=q3-q1
df3=df2[((df2>=q1-1.5*IQR)&(df2<=q3+1.5*IQR)).all(axis=1)]
```

```python
df3
```

```python
df3.boxplot()
```

```python
#outliers removed
# performing data analysis
df3.isnull().sum()
```
